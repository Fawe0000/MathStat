# Харьковцов А.В. Теория вероятностей и математическая статистика ДЗ_3

```python
import numpy as np
from math import factorial

# Формула числа сочетаний`
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

# Задача 1

Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.



```python
# Решение:
zp_v=np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
n=len(zp_v)
zp_v.sort()
m = zp_v.sum()/n
print (f"Среднее арифметическое зарплат выпускников составит: {m:.2f}")
s = (((zp_v-m)**2).sum()/(n-1))**.5
print (f"Среднее квадратичное отклонение зарплат выпускников составит: {s:.2f}")
s2sm = ((zp_v-m)**2).sum()/n
print (f"Смещенная дисперсия зарплат выпускников составит: {s2sm:.2f}")
s2nsm = ((zp_v-m)**2).sum()/(n-1)
print (f"Несмещенная дисперсия зарплат выпускников составит: {s2nsm:.2f}")
```

# Задача 2

В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?


Решение: 

Т.е. возможные варианты:
- из первой корзины 2 черных + из второй 1 черный и 3 белых;
- из первой корзины 1 белый и 1 черный + из второй 2 черных и 2 белых;
- из первой корзины 2 белых + из второй 3 черных и 1 белый

$$P=\frac{C_{3}^{2}}{C_{8}^{2}}*\frac{C_{7}^{1}*C_{5}^{3}}{C_{12}^{4}} + \frac{C_{5}^{1}*C_{3}^{1}}{C_{8}^{2}}*\frac{C_{7}^{2}*C_{5}^{2}}{C_{12}^{4}} + \frac{C_{5}^{2}}{C_{8}^{2}}*\frac{C_{7}^{3}*C_{5}^{1}}{C_{12}^{4}}\approx 36.87\%$$



```python
p1 = combinations(3, 2)/combinations(8, 2)*combinations(7, 1)*combinations(5, 3)/combinations(12, 4)
p2 = combinations(5, 1)*combinations(3, 1)/combinations(8, 2)*combinations(7, 2)*combinations(5, 2)/combinations(12, 4)
p3 = combinations(5, 2)/combinations(8, 2)*combinations(7, 3)*combinations(5, 1)/combinations(12, 4)
print (f"Вероятность того, что ровно три мяча белые составит примерно: {(p1 + p2 + p3)*100:.2f}%")
```

# Задача 3

На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.


Решение: 
Используем фокрмулу Байеса:
$$P(B\:|\:A) = \frac{P(B) \cdot P(A\:|\:B)}{P(A)}$$

Вероятность А, т.е. того, что попал хотябы один из них = 1/3 * 0,9 + 1/3 * 0,8 + 1/3 * 0,6 = 0,7667

Вероятность того, что попал первый: $$P(1\:|\:А) = \frac{1/3 * 0,9}{0,776} * 100\% \approx 39.13\% $$

Вероятность того, что попал второй: $$P(2\:|\:А) = \frac{1/3 * 0,8}{0,776} * 100\% \approx 34.78\% $$
Вероятность того, что попал третий: $$P(3\:|\:А) = \frac{1/3 * 0,6}{0,776} * 100\% \approx 26.08\% $$

Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?





Решение: 
Количество студентов на факультете А = 1/4
Количество студентов на факультете B = 1/4
Количество студентов на факультете C = 2/4

Вероятность того, что учится на факультете А = 1/4*0,8*100* = 20,00%
Вероятность того, что учится на факультете B = 1/4*0,7*100* = 17,50%

в) вероятность "хотябы одна" = 1 - "ни одной" = 1 - 0.9 * 0.8 * 0.75 = 46%
