```python
import numpy as np
from math import factorial
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

Из колоды в 52 карты извлекаются случайным образом 4 карты. 
a) Найти вероятность того, что все карты – крести.
б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

Решение а):
1) Находим число сочетаний из 52-х карт по 4:
ЧС4 = 52!*/(4!(52-4)!) = 52!/4!*48! = 270'725
2) Находим число сочетаний 4-х карт "крести". Т.е. нам нужно узнать, сколько вариантов есть сложить 4 "крести" из 13 "крести"
ЧС4к = 13!/(4!(13-4)! = 715
3) вероятность того, что из 270'725 вариантов выборки 4-х карт из колоды будет выбран один из 715 вариантов "крести составляет:
P4крести = 715/ 270'725 * 100% = 0.26%

```python
P4крести = combinations(13, 4)/combinations(52, 4)*100
P4крести
print (f"вероятность, что среди 4-х карт будут все крести составит примерно: {P4крести:.2f}%")
```

Решение b):

1) Находим число сочетаний одной карты "туз": ЧС1т
2) Находим число сочетаний из 48-ми карт по 3: ЧС3
3) Находим число сочетаний с 1 тузом = ЧС1т * ЧС3
4) вероятность того, что окажется 1 туз = ЧС1т * ЧС3 /  ЧС4 из 270'725 вариантов выборки 4-х карт из колоды будет выбран один из 715 вариантов "крести составляет:
Аналогично рассчитываем вероятность, что выпадет 2 туза; потом - 3 туза; потом 4 туза
Полная вероятность будет суммой всех этих 4-х вероятностей.

```python
P1t = combinations(4, 1) * combinations(48, 3) / combinations(52,4) * 100
P2t = combinations(4, 2) * combinations(48, 2) / combinations(52,4) * 100
P3t = combinations(4, 3) * combinations(48, 1) / combinations(52,4) * 100
P4t = combinations(4, 4) * combinations(48, 0) / combinations(52,4) * 100
P1_4t = P1t + P2t + P3t + P4t

print (f"вероятность, что среди 4-х карт окажется хотя бы один туз составляет примерно: {P1_4t:.0f}%")
```

На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

Решение:
1) Если нажимающий знает, что нужно нажать 3 цифры:
- Находим число сочетаний 3 кнопки из 10
- С первой попытки - значит "благоприятный исход" = 1
Вероятность = 1 / число сочетаний 3 из 10


2) Если нажимающий не знает, сколько кнопок нужно нажать:
- в этом случае, нужно найти все возможные сочетания из 10 кнопок. Т.к. состояние кнопок может быть либо "нажата", либо "не нажата", то получаем двоичный код. А число возможных комбинаций составит 2 в степени 10, за вычетом "нулевой комбинации" (человек не может ничего ни нажать).
- Предполагаем, что человек с первой попытки может попытаться нажать как одну кнопку, так и все 10. Тогда вероятность составит 1 / 1023

```python
print (f"Если человек будет пробовать нажимать только по 3 кнопки, тогда")
print (f"вероятность открыть дверь с 1 попытки составляет примерно: {(1/combinations(10, 3))*100:.4f}%")
print ()
print (f"Если человек не знает, сколько кнопок нужно нажимать, тогда")
print (f"вероятность открыть дверь с 1 попытки составляет примерно: {(1/(2 ** 10 - 1))*100:.4f}%")
```

В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

Решение:
Вероятность того, что первая деталь будет окрашена: 9/15=3/5
Вероятность того, что при вытащеной первой окрашеной детали, вторая тоже будет окрашена: (3/5)*(8/14)=12/35
Вероятность того, что после первых двух окрашенных, третья тоже будет окрашена: (12/35)*(7/13)=84/455 = 0,1846

Ответ: Вероятность того, что все 3 извлеченные детали окрашены, составит примерно 18,5%

```python
combinations(9, 3)/combinations(15, 3)*100
```
